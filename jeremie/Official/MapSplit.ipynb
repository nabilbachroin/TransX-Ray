```python
import pandas as pd
import numpy as np
```

```python
detail = pd.read_csv('~/Workshop/X-Ray/TransX-Ray/jeremie/Official/Data_Entry_2017_v2020.csv')
```

```python
detail.head()
```

```python
labels = detail['Finding Labels'].str.split('|')
image_path = detail[['Image Index']]
```

```python
dict = {}
list1 = ['No Finding', 'Atelectasis', 'Cardiomegaly', 'Effusion', 'Infiltration', 'Mass', 'Nodule', 'Pneumonia',
         'Pneumothorax', 'Consolidation', 'Edema', 'Emphysema', 'Fibrosis',
         'Pleural_Thickening', 'Hernia']

for i in list1:
    dict[i] = []
print(dict)
print(len(list1))
```

```python
for i in labels:
    for j in list1:
        if j in i:
            dict[j].append(1)
        else:
            dict[j].append(0)
```

```python
labelss = pd.DataFrame(dict)
```

```python
labelss.head()
```

```python
full_dataframe = pd.concat([image_path, labelss], axis=1)
```

```python
full_dataframe.head()
```

```python
for col in full_dataframe:
    print(full_dataframe[col].unique())
print("Total Dataset: ", len(full_dataframe))
```

```python
with open('/home/cries/Workshop/X-Ray/TransX-Ray/jeremie/Official/train_val_list.txt') as f:
    train = [line.rstrip('\n') for line in f]
    
with open('/home/cries/Workshop/X-Ray/TransX-Ray/jeremie/Official/test_list.txt') as f:
    test = [line.rstrip('\n') for line in f]
```

```python
print("Train Split: ", len(train))
print("Test Split: ", len(test))
print("Total Split: ", len(train)+len(test))
```

```python
train_df = full_dataframe[full_dataframe['Image Index'].isin(train)].reset_index(drop=True)
test_df = full_dataframe[full_dataframe['Image Index'].isin(test)].reset_index(drop=True)
```

```python
print("Train DF: ", len(train_df))
print("Test DF: ", len(test_df))
print("Total DF: ", len(train_df)+len(test_df))
```
